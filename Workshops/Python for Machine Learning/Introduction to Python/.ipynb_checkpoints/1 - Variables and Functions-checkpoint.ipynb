## Variable Types

```python
# Variable types

n = 5     # integer
x = 1.0   # float
b = True  # boolean
h = 'Hello World'  # string
l = [1, 2, 3, 4]   # list
t = (1, 2, 3, 4)   # tuple
s = {1, 2, 3, 4}   # set
d = {1: 'One', 2: 'Two', 3: 'Three', 4: 'Four'}  # dictionary


#############################################################
#############################################################


print('Varible "n" type:', type(n))
print('n =', n, end='\n\n')

print('Variable "x" type:', type(x))
print('x =', x, end='\n\n')

print('Variable "b" type:', type(b))
print('b =', b, end='\n\n')

print('Variable "h" type:', type(h))
print('h =', h, end='\n\n')

print('Variable "l" type:', type(l))
print('l =', l, end='\n\n')

print('Variable "t" type:', type(t))
print('t =', t, end='\n\n')

print('Variable "s" type:', type(s))
print('s =', s, end='\n\n')

print('Variable "d" type:', type(d))
print('d =', d)
```

## Arithmetic Operators



image source: https://www.w3schools.com/python/gloss_python_arithmetic_operators.asp

```python
# Arithmetic Operators

a = 20
b = 3

print("a =", a)
print("b =", b)
print()

print("a + b =", a + b)    # Addition Operator
print("a - b =", a - b)    # Substraction Operator
print("a * b =", a * b)    # Multiplication Operator
print("a / b =", a / b)    # Division Operator
print("a // b =", a // b)  # Floor Division Operator
print("a % b =", a % b)    # Modulus Operator
print("a ** b =", a ** b)  # Exponentiation Operator
```

## Comparison Operators



Image source: https://www.w3schools.com/python/gloss_python_comparison_operators.asp

```python
# Comparison Operators

x = 5.3
y = 5.0

print("x =", x)
print("y =", y)
print()

print("x == y:", x == y)   # Equal
print("x != y:", x != y)   # Not equal
print("x > y :", x > y)    # Greater than
print("x < y :", x < y)    # Less than
print("x >= y:", x >= y)   # Greater than or equal to
print("x <= y:", x <= y)   # Less than or equal to
```

## Logical Operators



Image source: https://www.w3schools.com/python/gloss_python_logical_operators.asp

```python
x = 6

print("x < 5 and x < 10:", x < 5 and x < 10)
print("x < 5 or x < 4:", x < 5 or x < 4)
print("not(x < 5 and x < 10):", not(x < 5 and x < 10))
```

## String Indexing

```python
string = "Hello World"

print("string =", string)
print("string[0] =", string[0])        # Element with index 0 (first element)
print("string[-1] =", string[-1])      # Last element
print("string[0:6] =", string[0:6])    # From the first to the 7th element (not included)
print("string[:6] =", string[:6])      # Equivalent to string[0:6]
print("string[6:] =", string[6:])      # From the 7th element to the end of the string
print("string[:] =", string[:])        # From the start to the end of the string
print("string[::2] =", string[::2])    # From the start to the end with step = 2
print("string[::-1] =", string[::-1])  # Reversed string
```

## String methods

```python
print("string.upper() =", string.upper())
print("string.lower() =", string.lower())
print("string.replace('l', 'r') =", string.replace('l', 'r'))
print("string.split(' ') =", string.split(' '))
```

For more string methods: https://www.w3schools.com/python/python_ref_string.asp

## Functions

```python
# A function that takes two numbers as parameters and returns the sum
def add(a, b):
    return a + b

result = add(5, 3)
print("Result =", result)
```

```python
# A function that takes two numbers as parameters and returns the product
def multiply(a, b):
    return a * b

result = multiply(5, 3)
print("Result =", result)
```

```python
# A function is a variable type
print(type(add))
```

```python
# Python magic!

function = add

result = function(5, 3)
print("Result =", result)
```

```python
# A function that takes a function as a parameter
def operation(a, b, func):
    return func(a, b)

result = operation(5, 3, add)
print("Result =", result)
```

```python
# Lambda function
add2 = lambda a, b: (a + b)    # return the sum of a and b

result = add2(5, 3)
print("Result =", result)
```
