```python
n = 5
premier = True


if (n % 2) == 0 and (n != 2):
    premier = False
    
for i in range(3, n//2, 2):
    if (n % i) == 0:
        premier = False
        break

if n == 1:
    print("n'est pas premier")
elif not premier:
    print("n'est pas premier")
else:
    print("premier")
```

```python
def est_premier(n):
    premier = True


    if (n % 2) == 0 and (n != 2):
        premier = False

    for i in range(3, n//2, 2):
        if (n % i) == 0:
            premier = False
            break

    if n == 1:
        print("n'est pas premier")
    elif not premier:
        print("n'est pas premier")
    else:
        print("premier")
```

```python
for i in range(1, 1000):
    print(i, end=' ')
    est_premier(i)
```

```python
numbers_list = [2, 4, 1, 3, 5, 4, 9, 2, 12, 35, 3]
for i in numbers_list:
    print(i, end=' ')
    est_premier(i)
```
